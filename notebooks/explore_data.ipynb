### Show Dataset Examples

```python
from IPython.display import display, Audio
import random

def show_data_examples(dataset, num_examples=5):
    """
    Hiển thị một số ví dụ audio + transcript từ dataset HuggingFace.
    
    Args:
        dataset: Dataset HuggingFace, có cột 'audio' và 'transcription'
        num_examples: Số lượng ví dụ muốn hiển thị
    """
    # Lấy ngẫu nhiên chỉ số ví dụ
    indices = random.sample(range(len(dataset)), num_examples)
    
    for idx in indices:
        example = dataset[idx]
        
        print(f"Example {idx}:")
        print(f"Transcript: {example['transcription']}")
        # Hiển thị player nghe audio
        display(Audio(example['audio']['array'], rate=example['audio']['sampling_rate']))
        print("-" * 50)
```

```python
show_data_examples(dataset)
```
