```python
from datasets import load_dataset, concatenate_datasets, Dataset
from datasets import Audio
import glob, os
import gc

def load_callfile_dataset(callfile_folders, label: str, cache_dir: str) -> Dataset:
    datasets_list = []

    for folder in callfile_folders:
        arrow_files = glob.glob(os.path.join(folder, "*.arrow"))
        callfile = load_dataset("arrow", data_files=arrow_files, split='train', cache_dir=cache_dir)
        callfile = callfile.cast_column("audio", Audio(sampling_rate=16000))

        cls_column = [f"{label}_callfile"] * len(callfile)
        callfile = callfile.add_column("cls", cls_column)

        datasets_list.append(callfile)

    concatenated = concatenate_datasets(datasets_list)

    del datasets_list, callfile, cls_column
    gc.collect()

    return concatenated
```
